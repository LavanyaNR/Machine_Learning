```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Simulated dataset (Area in square feet vs House Price in $1000s)
area = np.array([500, 700, 900, 1100, 1300, 1500, 1800, 2100, 2500, 3000]).reshape(-1, 1)
price = np.array([50, 65, 78, 90, 105, 115, 140, 160, 200, 250])  # Non-linear trend
```

```python
plt.scatter(area, price, color='blue', label='Actual Prices')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($1000s)')
plt.title('House Price vs Area')
plt.legend()
plt.show()
```

```python
# Train a simple Linear Regression model
linear_model = LinearRegression()
linear_model.fit(area, price)
price_pred_linear = linear_model.predict(area)

# Plot the linear regression line
plt.scatter(area, price, color='blue', label='Actual Prices')
plt.plot(area, price_pred_linear, color='red', label='Linear Regression')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($1000s)')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

```python
# Convert to polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
area_poly = poly.fit_transform(area)

# Train Polynomial Regression model
poly_model = LinearRegression()
poly_model.fit(area_poly, price)
price_pred_poly = poly_model.predict(area_poly)

# Plot Polynomial Regression results
plt.scatter(area, price, color='blue', label='Actual Prices')
plt.plot(area, price_pred_poly, color='green', label='Polynomial Regression (Degree 2)')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($1000s)')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Compute performance metrics
linear_mse = mean_squared_error(price, price_pred_linear)
poly_mse = mean_squared_error(price, price_pred_poly)
linear_r2 = r2_score(price, price_pred_linear)
poly_r2 = r2_score(price, price_pred_poly)

print(f"Linear Regression MSE: {linear_mse:.2f}, R²: {linear_r2:.2f}")
print(f"Polynomial Regression MSE: {poly_mse:.2f}, R²: {poly_r2:.2f}")
```
